```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
%matplotlib inline
x =[3,1,2,-1,4,6]

h = [3,2,1]
print (x)
print (h)
```

```python
n=np.arange(0,(len(x)+len(h)-1),1)
```

```python
y = signal.convolve(x,h)
print (y)
plt.stem(n,y)
plt.xlabel("Discrete Time Signal")
plt.ylabel("Amplitude")
plt.title(r'Convolution Sum of  $y[n]=\sum_{k=0}^\infty x[k]h[n-k]$')

#plot(signal.convolve(a, b));
```

```python
fig = figure()
ax = fig.add_subplot(111)
t = linspace(-1,1,100) 
fs=5.0 # sampling rate
k=array(sorted(set((t*fs).astype(int)))) # sorted coefficient list
fig=figure()
ax = fig.add_subplot(111)
ax.plot(t,(sin(2*pi*(k[:,None]/fs))*sinc(k[:,None]-fs*t)).T,'--', # individual whittaker functions
        t,(sin(2*pi*(k[:,None]/fs))*sinc(k[:,None]-fs*t)).sum(axis=0),'k-', # whittaker interpolant
     k/fs,sin(2*pi*k/fs),'ob')# samples
ax.set_xlabel('time',fontsize=14)
ax.axis((-1.1,1.1,-1.1,1.1));
ax.figure.savefig('figure_001.png', bbox_inches='tight', dpi=300)
```

```python
fig = figure()
ax = fig.add_subplot(111) # create axis handle
k=0
fs=2 # makes this plot easier to read
ax.plot (t,sinc( k - fs * t), 
         t,sinc( k+1 - fs * t),'--',k/fs,1,'o',(k)/fs,0,'o',
         t,sinc( k-1 - fs * t),'--',k/fs,1,'o',(-k)/fs,0,'o'
)
ax.hlines(0,-1,1)
ax.vlines(0,-.2,1)
ax.annotate('sample value goes here',
            xy=(0,1),
            xytext=(-1+.1,1.1),
            arrowprops={'facecolor':'red','shrink':0.05},
            )
ax.annotate('no interference here',
            xy=(0,0),
            xytext=(-1+.1,0.5),
            arrowprops={'facecolor':'green','shrink':0.05},
            )
ax.figure.savefig('figure_00@.png', bbox_inches='tight', dpi=300)
```

```python
from scipy import signal
sig = np.repeat([0., 1., 1., 0., 1., 0., 0., 1.], 128)
sig_noise = sig + np.random.randn(len(sig))
corr = signal.correlate(sig_noise, np.ones(128), mode='same') / 128
import matplotlib.pyplot as plt
clock = np.arange(64, len(sig), 128)
fig, (ax_orig, ax_noise, ax_corr) = plt.subplots(3, 1, sharex=True)
ax_orig.plot(sig)
ax_orig.plot(clock, sig[clock], 'ro')
ax_orig.set_title('Original signal')
ax_orig.annotate('clock signal',
            xy=(64,0),
            xytext=(0,1.2),
            arrowprops={'facecolor':'red','shrink':0.05},
            )
ax_orig.annotate('clock signal',
            xy=(192,1),
            xytext=(192,0.5),
            arrowprops={'facecolor':'green','shrink':0.05},
            )
ax_noise.plot(sig_noise)
ax_noise.set_title('Signal with noise')
ax_corr.plot(corr)
ax_corr.plot(clock, corr[clock], 'ro')
ax_corr.axhline(0.5, ls=':')
ax_corr.set_title('Cross-correlated with rectangular pulse')
ax_orig.margins(0, 0.1)
fig.tight_layout()
fig.show()
fig.savefig('Correlation_adaptivefilter.png', bbox_inches='tight', dpi=300)
```
